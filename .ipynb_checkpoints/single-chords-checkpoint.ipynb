```python
chord_df = pd.read_csv("./chords.csv", dtype={'frets': object})
chord_df.head()
```

```python
for i in chord_df.index:
    name = chord_df['name'][i]
    frets = chord_df['frets'][i]
    c = uchord.Chord(name, frets).to_svg()

    with open("../chords-v2/{0}.svg".format(name), 'w') as f:
        f.write(c)
        print("wrote {0}".format(name))
        f.close()
        
    display(SVG(c))
```
